### Evaluate on Test Set

After training completes, run the cell below to evaluate the trained model on the test set and see final metrics.

```python
# NOT exported - for interactive dev only
from pytorch_lightning import Trainer
from src.datamodules.mnist_datamodule import MNISTDataModule
from src.modules.mnist_litmodule import MNISTLitModule
from src.models.simpledense import SimpleDenseNet
from src.run_utils import find_best_checkpoint

# Load the best checkpoint
model, error = find_best_checkpoint(SimpleDenseNet, MNISTLitModule)

if model:
    # Run test
    datamodule = MNISTDataModule()
    trainer = Trainer(accelerator="cpu")
    results = trainer.test(model, datamodule=datamodule)
    print(f"\nTest Results: {results}")
else:
    print(f"Error: {error}")
```
